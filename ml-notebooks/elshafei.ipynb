```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
summary = pd.read_excel('data/elshafei.xlsx', sheet_name='Summary').set_index('No. Participant')
```

```python
df= pd.read_excel('data/elshafei.xlsx', sheet_name='P1')
df.columns = [c.strip() for c in df.columns]
df['time'] = np.arange(df.shape[0]) / 50
for c in ['ax', 'ay', 'az']:
    df[c] /= 16384.0
for c in ['gx', 'gy', 'gz']:
    df[c] /= 131.0
df.shape
```

```python
fig, axes = plt.subplots(3, 1, figsize=(10, 7), sharex=True)

dfplot = df#[df['time'].between(20,50)]

axes[0].plot(dfplot.time, dfplot[['ax', 'ay', 'az']], label=['acc_x', 'acc_y', 'acc_z'])
axes[1].plot(dfplot.time, dfplot[['gx', 'gy', 'gz']], label=['gyr_x', 'gyr_y', 'gyr_z'])
axes[2].plot(dfplot.time, dfplot[['yaw', 'pitch', 'roll']], label=['yaw', 'pitch', 'roll'])

axes[0].set_ylabel('Acceleration')
axes[1].set_ylabel('Gyroscope')
axes[2].set_ylabel('Position')
axes[2].set_xlabel('Time (s)')
[ax.legend(loc=(1.05, 0)) for ax in axes]
```
